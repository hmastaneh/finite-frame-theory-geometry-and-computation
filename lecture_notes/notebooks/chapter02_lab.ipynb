## Lab Exercise 2.5 (Optional, Challenge): How Many Samples Are Enough?

Using the triangular frame ($A=B=1.5$), run `estimate\_frame\_bounds` with `num\_samples` at $10, 100, 1000, 10000, 100000$. Record errors $|A_{\text{est}}-1.5|$, $|B_{\text{est}}-1.5|$. Plot error vs. num\_samples on log-log axes.

```python
f1 = np.array([1.0, 0.0])
f2 = np.array([-0.5, np.sqrt(3)/2])
f3 = np.array([-0.5, -np.sqrt(3)/2])
triangle_vectors = [f1, f2, f3]
A_true = B_true = 1.5

sample_sizes = [10, 100, 1000, 10_000, 100_000]
n_repeats = 5

A_errors = []
B_errors = []

for num_samples in sample_sizes:
    a_errs, b_errs = [], []
    for rep in range(n_repeats):
        A_est, B_est = None, None  # TODO: estimate_frame_bounds(triangle_vectors, n=2, num_samples=num_samples, seed=rep)
        a_errs.append(abs(A_est - A_true))
        b_errs.append(abs(B_est - B_true))
    A_errors.append(np.mean(a_errs))
    B_errors.append(np.mean(b_errs))

plt.figure(figsize=(7, 5))
plt.loglog(sample_sizes, A_errors, 'o-', label='error in A')
plt.loglog(sample_sizes, B_errors, 's-', label='error in B')
plt.xlabel('num_samples')
plt.ylabel('mean absolute error')
plt.title('Convergence of random-sampling frame bound estimates')
plt.legend()
plt.grid(True, which='both', alpha=0.3)
plt.show()
```
